```python
import pandas as pd
```

```python
df = pd.read_csv('data/data.csv', index_col=0)
df
```

```python
# for every column except the first two, create the shift of 1 to 6
for col in df.columns[2:]:
    for i in range(1, 7):
        df[f'{col}_lag_{i}'] = df[col].shift(i)
df.dropna(inplace=True)
df
```

```python
df.to_csv('data/data_lagged.csv')
```
